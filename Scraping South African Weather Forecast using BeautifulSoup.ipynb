### Import Modules

```python
import pandas as pd
import numpy as np
import csv
import requests
from bs4 import BeautifulSoup as bs
```

### Requests page source from url

```python
url = "https://www.weathersa.co.za/"
```

```python
page = requests.get(url)
page
```

```python
#Display page source code

page.content
```

```python
soup = bs(page.content, "html.parser")

soup.prettify()
```

```python
#getting the table

weather_table = soup.find("div", class_="col-sm-7 topCities")

headers = []

for i in weather_table.find_all("th"):
    title = i.text
    headers.append(title)
```

```python
headers
```

```python
df = pd.DataFrame(columns=headers)
df
```

```python
for row in weather_table.find_all('tr')[1:]:
    data = row.find_all('td')
    row_data = [td.text.strip() for td in data]
    length = len(df)
    df.loc[length] = row_data
```
